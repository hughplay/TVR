```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Palatino Linotype')
# plt.rc('xtick')
# plt.rc('ytick')
plt.rcParams['font.size'] = 18
plt.rcParams['axes.linewidth'] = 2

fig = plt.figure(figsize=(7, 4.5))
ax = fig.add_subplot(1, 1, 1)

x = [10, 50, 100, 200, 300, 400, 500]
x_basic = [10, 50, 125]
y_basic = [0.7710, 0.9530, 0.9744]
y_event = [0.1701, 0.3234, 0.4246, 0.5340, 0.5892, 0.6245, 0.6553]
y_view = [0.0593, 0.2034, 0.3144, 0.4452, 0.5101, 0.5473, 0.6012]

# yp = [0.2071, 0.3787, 0.4334, 0.4823, 0.5086, 0.5280, 0.5416]

ax.plot(x_basic, y_basic, '^', markersize=9, color='k', ls='solid', linewidth=2, label='basic')
ax.plot(x, y_event, 'o', markersize=7, color='k', ls='solid', linewidth=2, label='event')
ax.plot(x, y_view, 'o', markersize=7, color='k', ls='dashed', linewidth=2, label='view')

ax.legend(loc='lower right', fontsize=16, edgecolor='white', fancybox=False)

ax.set_xlabel('Training Samples (k)')
ax.set_ylabel('Acc')


# zip joins x and y coordinates in pairs
for _x, _y in zip(x, y_basic):

    label = "{:.4f}".format(_y)

    if _x < 50:
        xytext = (35, -15)
    elif _x < 100:
        xytext = (25, -25)
    else:
        xytext = (40, -25)

    plt.annotate(label, # this is the text
                 (_x, _y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=xytext, # distance from text to points (x,y)
                 ha='center', fontsize=16) # horizontal alignment can be left, right or center

for _x, _y in zip(x, y_event):

    label = "{:.4f}".format(_y)


    if _x == 10:
        xytext = (5, 35)
    elif _x < 100:
        xytext = (0, 20)
    elif _x < 200:
        xytext = (5, 15)
    else:
        xytext = (-10, 15)

    plt.annotate(label, # this is the text
                 (_x, _y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=xytext, # distance from text to points (x,y)
                 ha='center', fontsize=16) # horizontal alignment can be left, right or center

for _x, _y in zip(x, y_view):

    label = "{:.4f}".format(_y)

    if _x == 10:
        xytext = (30, -10)
    elif _x < 200:
        xytext = (30, -20)
        
    elif _x <= 400:
        xytext = (10, -30)
    else:
        xytext = (-10, -30)

    plt.annotate(label, # this is the text
                 (_x, _y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=xytext, # distance from text to points (x,y)
                 ha='center', fontsize=16, color='k') # horizontal alignment can be left, right or center

_ = plt.xticks(x, fontsize=16)
_ = plt.yticks(np.arange(0.,1.1,0.1), fontsize=16)
plt.savefig('size.pdf', bbox_inches='tight', pad_inches=0.05, dpi=300)
```
